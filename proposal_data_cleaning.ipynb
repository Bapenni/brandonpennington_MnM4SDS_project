```python
import geopandas as gpd
import matplotlib.pyplot as plt
print("Load Complete")
```

```python
crimes_2020 = gpd.read_file("https://opendata.arcgis.com/datasets/f516e0dd7b614b088ad781b0c4002331_2.geojson")
crimes_2019 = gpd.read_file("https://opendata.arcgis.com/datasets/f08294e5286141c293e9202fcd3e8b57_1.geojson")
crimes_2018 = gpd.read_file("https://opendata.arcgis.com/datasets/38ba41dd74354563bce28a359b59324e_0.geojson")
crimes_2017 = gpd.read_file("https://opendata.arcgis.com/datasets/6af5cb8dc38e4bcbac8168b27ee104aa_38.geojson")
police_sectors = gpd.read_file("https://opendata.arcgis.com/datasets/6ac17c2ff8cc4e20b3768dd1b98adf7a_23.geojson")
police_stations = gpd.read_file("https://opendata.arcgis.com/datasets/9e465c1e6dfd4605a7632ed5737644f3_11.geojson")
dc_atms = gpd.read_file("https://opendata.arcgis.com/datasets/5555b90602a94d9c85d63ee62a860c6e_31.geojson")
# metro_entrances = gpd.read_file("https://opendata.arcgis.com/datasets/7883e07b7b5d4b74b0b5ea4f02f895af_63.geojson")
metro_entrance_points = gpd.read_file("https://opendata.arcgis.com/datasets/ab5661e1a4d74a338ee51cd9533ac787_50.geojson")
metro_lines = gpd.read_file("https://opendata.arcgis.com/datasets/a29b9dbb2f00459db2b0c3c56faca297_106.geojson")
print("Download complete")
```

```python
# crimes_2020.shape,crimes_2019.shape,crimes_2018.shape,crimes_2017.shape,police_sectors.shape,police_stations.shape,dc_atms.shape,
crimes_2020.shape[0] + crimes_2019.shape[0] + crimes_2018.shape[0] + crimes_2017.shape[0]
```

```python
fig, ax = plt.subplots(figsize=(18,16))
police_sectors.plot(ax=ax)
dc_atms.plot(ax=ax, color='black', markersize=12.0, label='ATMs')
crimes_2020.plot(ax=ax, markersize=0.4, color='red', label='crimes')
crimes_2019.plot(ax=ax, markersize=0.4, color='red')
crimes_2018.plot(ax=ax, markersize=0.4, color='red')
crimes_2017.plot(ax=ax, markersize=0.4, color='red')
metro_entrance_points.plot(ax=ax, color='yellow', label='Metro Entrances')
metro_lines.plot(ax=ax, color='grey', label='Metro Lines')
police_stations.plot(ax=ax, color='pink', markersize=75.0, marker='+', label='Police Stations')
leg = ax.legend();
```

```python
crime_total_time = crimes_2020.SHIFT.value_counts() + crimes_2019.SHIFT.value_counts() + crimes_2018.SHIFT.value_counts() + crimes_2017.SHIFT.value_counts()
crime_total_time
```

```python
totals = pd.DataFrame(crime_total_time)
totals.reset_index()
plt.bar(totals.index, totals.SHIFT)
```
